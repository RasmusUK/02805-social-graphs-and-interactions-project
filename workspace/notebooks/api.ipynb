```python
# Install Required Libraries

# Import Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from datetime import datetime

# STEP 1: Data Collection
def fetch_html(user_id):
    """
    Fetch the HTML content of a user's Stack Overflow profile page.
    """
    url = f"https://stackoverflow.com/users/{user_id}"
    response = requests.get(url)
    response.raise_for_status()
    return BeautifulSoup(response.content, "html.parser")


def fetch_users(page=1):
    url = "https://api.stackexchange.com/2.3/users"
    params = {
        "site": "stackoverflow",
        "order": "desc",
        "sort": "reputation",
        "pagesize": 100,
        "page": page
    }
    response = requests.get(url, params=params)
    print(f"Status Code: {response.status_code}")
    print(f"Response Text: {response.text[:500]}")  # Print first 500 characters of the response

    # If the response is empty or invalid
    if response.status_code != 200 or not response.text:
        raise Exception("Failed to fetch users from the API")

    data = response.json()
    if "items" not in data:
        print(f"Error fetching users: {data}")
        return {"items": []}

    return data


def fetch_user_tags(user_id):
    """
    Fetch tags for a given user from Stack Overflow using the API.
    """
    url = f"https://api.stackexchange.com/2.3/users/{user_id}/tags"
    params = {
        "site": "stackoverflow",
        "pagesize": 100
    }
    response = requests.get(url, params=params)
    data = response.json()

    # Check for errors
    if "items" not in data:
        print(f"Error fetching tags for user {user_id}: {data}")
        return {"items": []}

    return data


def fetch_user_posts(user_id, post_type="questions"):
    """
    Fetch questions or answers posted by a user.
    """
    url = f"https://api.stackexchange.com/2.3/users/{user_id}/{post_type}"
    params = {
        "site": "stackoverflow",
        "order": "desc",
        "sort": "creation",
        "pagesize": 100
    }
    response = requests.get(url, params=params)
    data = response.json()

    # Check for errors
    if "items" not in data:
        print(f"Error fetching posts for user {user_id}: {data}")
        return {"items": []}

    return data


# Fetch sample data for demonstration
users_data = fetch_users(page=1)
user_tags = fetch_user_tags(user_id=6309)  # Replace with a sample user ID
user_posts = fetch_user_posts(user_id=6309, post_type="questions")  # Replace with a sample user ID

# STEP 2: Data Preprocessing
def preprocess_tags(tags_data, user_id):
    """
    Preprocess tag data into a pandas DataFrame and include user_id.
    """
    if "items" not in tags_data or not tags_data["items"]:
        print(f"No tags found for user {user_id}")
        return pd.DataFrame()  # Return an empty DataFrame

    tags_df = pd.DataFrame(tags_data["items"])
    tags_df = tags_df[["name", "count"]]
    tags_df.rename(columns={"name": "tag", "count": "tag_count"}, inplace=True)
    tags_df["user_id"] = user_id  # Add user_id as a column
    return tags_df


# Preprocess Tags for a Single User
tags_df = preprocess_tags(user_tags, user_id=6309)
print(tags_df.head())

# STEP 3: Build the Network
def build_bipartite_network(users_data, tags_df):
    """
    Build a bipartite network from user and tag data.
    """
    B = nx.Graph()

    # Add user nodes
    for user in users_data["items"]:
        B.add_node(user["user_id"], bipartite=0, reputation=user.get("reputation", 0))

    # Add tag nodes and edges
    for _, row in tags_df.iterrows():
        B.add_node(row["tag"], bipartite=1)  # Add tag node
        B.add_edge(row["user_id"], row["tag"], weight=row["tag_count"])  # Add edge with weight

    return B


# Create the Bipartite Network
bipartite_network = build_bipartite_network(users_data, tags_df)

# Project to User-to-User Network
user_network = nx.bipartite.projected_graph(
    bipartite_network,
    [n for n, d in bipartite_network.nodes(data=True) if d["bipartite"] == 0]
)

# STEP 4: Analyze and Visualize
def analyze_network(network):
    """
    Analyze the network by computing basic metrics.
    """
    degree_centrality = nx.degree_centrality(network)
    clustering = nx.clustering(network)

    print("Top Users by Degree Centrality:")
    print(sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10])

    print("\nTop Users by Clustering Coefficient:")
    print(sorted(clustering.items(), key=lambda x: x[1], reverse=True)[:10])


# Analyze the User-to-User Network
analyze_network(user_network)


def visualize_network(network):
    """
    Visualize the network.
    """
    plt.figure(figsize=(10, 8))
    nx.draw(
        network,
        with_labels=True,
        node_size=50,
        font_size=8,
        alpha=0.7,
        edge_color="gray"
    )
    plt.title("User-to-User Network")
    plt.show()


# Visualize the User-to-User Network
visualize_network(user_network)


# STEP 5: Interpret Findings
# Example: Identify clusters and central users
def detect_communities(network):
    """
    Detect communities in the network using the Louvain method.
    """
    from networkx.algorithms.community import greedy_modularity_communities
    communities = list(greedy_modularity_communities(network))
    print(f"Number of communities detected: {len(communities)}")
    return communities


# Detect communities in the user network
user_communities = detect_communities(user_network)
```

```python
# Fetch multiple pages of users
all_users = []
for page in range(1, 6):  # Adjust the range for more users
    data = fetch_users(page=page)
    all_users.extend(data["items"])

# Collect tags for all users
all_user_tags = []
for user in all_users:
    user_id = user["user_id"]
    user_tags = fetch_user_tags(user_id)
    tags_df = preprocess_tags(user_tags, user_id=user_id)
    all_user_tags.append(tags_df)

# Combine tag data into a single DataFrame
tags_df_combined = pd.concat(all_user_tags, ignore_index=True)
print(tags_df_combined.head())
```
